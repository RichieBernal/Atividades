***Import the necessary library***

```python
import pandas as pd
```

### **Read data**

**Knowning original columns**

```python
# Create a data Frame from .csv separated by semicolon
df = pd.read_csv('Absenteeism_at_work.csv', sep=';')

# Print the current column names
print("Original column names:")
print(df.columns)
```

**# Define a dictionary with the current column names as keys and new column names as values**

```python
column_names = {
    'D': 'D_new',
    'Reason for absence': 'Reason_for_absence_new',
    'Month of absence': 'Month_of_absence_new',
    'Day of the week': 'Day_of_the_week_new',
    'Seasons': 'Seasons_new',
    'Transportation expense': 'Transportation_expense_new',
    'Distance from Residence to Work': 'Distance_from_Residence_to_Work_new',
    'Service time': 'Service_time_new',
    'Age': 'Age_new',
    'Work load Average/day': 'Work_load_Average_per_day_new',
    'Disciplinary failure': 'Disciplinary_failure_new',
    'Education': 'Education_new',
    'Son': 'Son_new',
    'Social drinker': 'Social_drinker_new',
    'Social smoker': 'Social_smoker_new',
    'Pet': 'Pet_new',
    'Weight': 'Weight_new',
    'Height': 'Height_new',
    'Body mass index': 'Body_mass_index_new',
    'Absenteeism time in hours': 'Absenteeism_time_in_hours_new'
}
```

#  **Rename the columns using the dictionary**

```python
df.rename(columns=column_names, inplace=True)
```

```python
# Print the new column names
print("\nNew column names:")
df.head()
```

# **Using labels**

**Get the list of column names from the DataFrame**

```python
labels = list(df.columns)
```

**Iterate over the list of column names**

```python
for i in range(len(labels)):
    # Replace spaces with underscores and convert to lower case
    labels[i] = labels[i].replace(' ', '_').lower()
```

**Assign the modified column names back to the DataFrame**

```python
df.columns = labels
```
